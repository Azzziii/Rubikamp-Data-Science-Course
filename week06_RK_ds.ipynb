```python
def title_organizer(x):
    x = x.lower()
    if ('data scientist' in x) or ('data science' in x) :
        return 'data scientist'
    elif 'data engineer' in x:
        return 'data engineer'
    elif 'deep learning' in x:
        return 'deep learning engineer'
    elif 'machine learning' in x:
        return 'machine learning engineer'
    elif 'director' in x:
        return 'director'
    elif 'manager' in x:
        return 'manager'
    elif 'analyst' in x:
        return 'analyst'
    elif 'researcher' in x:
        return 'researcher'
    else:
        return 'other'
```

```python
'Data'.lower()
```

```python
title_organizer('Business Intelligence Analyst Senior (Chief Financial Office)')
```

```python
df['Job Title'].apply(title_organizer)
```

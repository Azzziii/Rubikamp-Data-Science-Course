```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Azzziii/Rubikamp-Data-Science-Course/main/df.csv')
```

```python
df
```

```python
df['Salary Estimate'][0].replace('$', ' ')
```

```python
def dollors_remover(a):
  a = a.replace('$', ' ')
  return a
```

```python
dollors_remover(df['Salary Estimate'][0])
```

```python
dollors_remover('fh$wb$h48$$99kd')
```

```python
dollors_remover('465$')
```

```python
dollors_remover('Azam Soltanshah')
```

```python
df['Salary Estimate'].apply(dollors_remover)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(dollors_remover)
```

```python
def k_remover(a):
  a = a.replace('K' , ' ')
  return a
```

```python
k_remover(df['Salary Estimate'][0])
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(k_remover)
```
